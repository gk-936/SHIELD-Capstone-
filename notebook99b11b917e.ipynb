```python
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import (
    classification_report, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay,
    RocCurveDisplay, f1_score,
    precision_score, recall_score
)
from sklearn.preprocessing import MinMaxScaler

!pip install pyod -q
from pyod.models.ecod   import ECOD
from pyod.models.copod  import COPOD
from pyod.models.hbos   import HBOS
from pyod.models.lof    import LOF


# ════════════════════════════════════════════════════════════════
# PHASE 2 — FEATURE SETS + ROBUST SCALING
# ════════════════════════════════════════════════════════════════

df = pd.read_parquet("/kaggle/input/datasets/gkd936/ransmap-features")
df = df.replace([np.inf, -np.inf], 0)

# ── Final feature sets ────────────────────────────────────────
STORAGE_FEATURES = [
    # Volume
    "total_accesses", "mean_access_size", "std_access_size",
    # Entropy original
    "mean_entropy", "std_entropy", "high_entropy_ratio",
    "entropy_spike_count",
    # Entropy new
    "entropy_trend", "entropy_variance_blocks",
    "peak_entropy_ratio",
    # Temporal original
    "duration_sec", "access_rate", "inter_access_mean",
    "inter_access_std", "burstiness",
    # Temporal new
    "io_acceleration",
    # Spatial
    "unique_blocks", "block_range", "sequential_ratio",
    # Cross-group
    "write_size_uniformity", "entropy_access_rate_ratio",
]

MEMORY_FEATURES = [
    "write_count", "write_ratio", "rw_ratio",
    "write_entropy_mean", "high_entropy_write_ratio",
    "write_acceleration",
]

ALL_FEATURES = STORAGE_FEATURES + MEMORY_FEATURES
print(f"Storage features : {len(STORAGE_FEATURES)}")
print(f"Memory features  : {len(MEMORY_FEATURES)}")
print(f"Total features   : {len(ALL_FEATURES)}")

# ── Train/Val/Test splits ─────────────────────────────────────
original = df[df.split == "original"]
extra    = df[df.split == "extra"]
mix      = df[df.split == "mix"]
variants = df[df.split == "variants"]

X_orig = original[ALL_FEATURES].values
y_orig = original["label"].values

X_train, X_val, y_train, y_val = train_test_split(
    X_orig, y_orig,
    test_size=0.2,
    stratify=y_orig,
    random_state=42
)

X_extra    = extra[ALL_FEATURES].values;    y_extra    = extra["label"].values
X_mix      = mix[ALL_FEATURES].values;      y_mix      = mix["label"].values
X_variants = variants[ALL_FEATURES].values; y_variants = variants["label"].values

s_idx = list(range(len(STORAGE_FEATURES)))
m_idx = list(range(len(STORAGE_FEATURES), len(ALL_FEATURES)))

benign_mask = (y_train == 0)

# ── RobustScaler (fit on benign train only) ───────────────────
scaler_s = RobustScaler()
scaler_m = RobustScaler()

scaler_s.fit(X_train[benign_mask][:, s_idx])
scaler_m.fit(X_train[benign_mask][:, m_idx])

def scale(X):
    Xs = scaler_s.transform(X[:, s_idx])
    Xm = scaler_m.transform(X[:, m_idx])
    return Xs, Xm

Xs_tr,  Xm_tr  = scale(X_train)
Xs_val, Xm_val = scale(X_val)
Xs_ex,  Xm_ex  = scale(X_extra)
Xs_mix, Xm_mix = scale(X_mix)
Xs_var, Xm_var = scale(X_variants)

# Combined scaled arrays
X_tr_all  = np.hstack([Xs_tr,  Xm_tr])
X_val_all = np.hstack([Xs_val, Xm_val])
X_ex_all  = np.hstack([Xs_ex,  Xm_ex])
X_mx_all  = np.hstack([Xs_mix, Xm_mix])
X_vr_all  = np.hstack([Xs_var, Xm_var])

benign_s_tr = Xs_tr[benign_mask]
benign_m_tr = Xm_tr[benign_mask]
benign_all  = X_tr_all[benign_mask]

print(f"\nTrain benign : {len(benign_all)}")
print(f"Train ransom : {benign_mask.sum()==False}")
print(f"Val          : {len(X_val_all)}")
print(f"Extra        : {len(X_ex_all)}")
print(f"Mix          : {len(X_mx_all)}")
print(f"Variants     : {len(X_vr_all)}")


# ════════════════════════════════════════════════════════════════
# PHASE 3 — TRAIN COUNCIL OF MODELS
# ════════════════════════════════════════════════════════════════

print("\nTraining council of models...")

# 1. IF Storage specialist
IF_storage = IsolationForest(
    n_estimators=500,
    max_samples="auto",
    max_features=0.8,
    contamination="auto",
    random_state=42,
    n_jobs=-1
)
IF_storage.fit(benign_s_tr)
print("✅ IF_storage trained")

# 2. IF Memory specialist
IF_memory = IsolationForest(
    n_estimators=500,
    max_samples="auto",
    max_features=0.8,
    contamination="auto",
    random_state=42,
    n_jobs=-1
)
IF_memory.fit(benign_m_tr)
print("✅ IF_memory  trained")

# 3. ECOD — full feature set
ecod = ECOD(contamination=0.05, n_jobs=-1)
ecod.fit(benign_all)
print("✅ ECOD trained")

# 4. COPOD — full feature set
copod = COPOD(contamination=0.05, n_jobs=-1)
copod.fit(benign_all)
print("✅ COPOD trained")

# 5. HBOS — full feature set
hbos = HBOS(contamination=0.05, n_bins=20)
hbos.fit(benign_all)
print("✅ HBOS trained")

# 6. LOF — full feature set
lof = LOF(
    n_neighbors=20,
    contamination=0.05,
    n_jobs=-1
)
lof.fit(benign_all)
print("✅ LOF trained")


# ════════════════════════════════════════════════════════════════
# PHASE 4 — SCORE NORMALIZATION + FUSION
# ════════════════════════════════════════════════════════════════

def get_raw_scores(Xs, Xm):
    """Get raw anomaly scores from all 6 models."""
    X_all = np.hstack([Xs, Xm])
    return {
        "IF_storage": -IF_storage.decision_function(Xs),
        "IF_memory":  -IF_memory.decision_function(Xm),
        "ECOD":        ecod.decision_function(X_all),
        "COPOD":       copod.decision_function(X_all),
        "HBOS":        hbos.decision_function(X_all),
        "LOF":         lof.decision_function(X_all),
    }

# Fit MinMax normalizer on TRAINING scores
train_raw = get_raw_scores(Xs_tr, Xm_tr)
normalizers = {}

for name, scores in train_raw.items():
    norm = MinMaxScaler()
    norm.fit(scores.reshape(-1, 1))
    normalizers[name] = norm

def get_normalized_scores(Xs, Xm):
    """Get per-model normalized scores (0-1 scale)."""
    raw  = get_raw_scores(Xs, Xm)
    norm = {}
    for name, scores in raw.items():
        norm[name] = normalizers[name].transform(
            scores.reshape(-1, 1)
        ).flatten()
    return norm


# ── Grid search best weights on validation ────────────────────
print("\nSearching best ensemble weights...")

model_names = ["IF_storage","IF_memory","ECOD","COPOD","HBOS","LOF"]
val_norm    = get_normalized_scores(Xs_val, Xm_val)

best_f1, best_weights, best_t = 0, None, 0

# Sample weight combinations
from itertools import product

weight_options = [0.1, 0.2, 0.3, 0.4]
combos = list(product(weight_options, repeat=6))
# Filter to combos that sum to ~1.0
combos = [w for w in combos if abs(sum(w) - 1.0) < 0.05]
print(f"Testing {len(combos)} weight combinations...")

for weights in combos:
    ensemble_scores = sum(
        weights[i] * val_norm[m]
        for i, m in enumerate(model_names)
    )
    for t in np.percentile(ensemble_scores, np.arange(50, 100, 1.0)):
        preds = (ensemble_scores > t).astype(int)
        f1    = f1_score(y_val, preds, zero_division=0)
        if f1 > best_f1:
            best_f1      = f1
            best_weights = weights
            best_t       = t

print(f"\nBest weights:")
for name, w in zip(model_names, best_weights):
    print(f"  {name:12s}: {w:.1f}")
print(f"Best threshold : {best_t:.6f}")
print(f"Best Val F1    : {best_f1:.4f}")


def ensemble_score(Xs, Xm):
    norm = get_normalized_scores(Xs, Xm)
    return sum(
        best_weights[i] * norm[m]
        for i, m in enumerate(model_names)
    )


# ════════════════════════════════════════════════════════════════
# PHASE 5 — TWO-THRESHOLD CALIBRATION
# ════════════════════════════════════════════════════════════════

val_scores = ensemble_score(Xs_val, Xm_val)

# HIGH alert  → maximize precision (fewer false alarms)
# MEDIUM alert → maximize recall (catch everything)
best_high_t, best_med_t = best_t, best_t
best_high_prec, best_med_recall = 0, 0

for t in np.percentile(val_scores, np.arange(50, 99, 0.5)):
    preds = (val_scores > t).astype(int)
    p = precision_score(y_val, preds, zero_division=0)
    r = recall_score(y_val, preds, zero_division=0)
    if p > best_high_prec and r > 0.5:
        best_high_prec = p
        best_high_t    = t
    if r > best_med_recall and p > 0.5:
        best_med_recall = r
        best_med_t      = t

print(f"\nThreshold calibration:")
print(f"  HIGH   alert threshold : {best_high_t:.6f}  (precision={best_high_prec:.3f})")
print(f"  MEDIUM alert threshold : {best_med_t:.6f}  (recall={best_med_recall:.3f})")


def predict_with_levels(scores, high_t, med_t):
    """
    Returns:
      2 = HIGH alert   (very likely ransomware)
      1 = MEDIUM alert (suspicious)
      0 = Benign
    """
    result = np.zeros(len(scores), dtype=int)
    result[scores > med_t]  = 1
    result[scores > high_t] = 2
    return result


# ════════════════════════════════════════════════════════════════
# PHASE 6 — EVALUATION
# ════════════════════════════════════════════════════════════════

def evaluate(name, Xs, Xm, y, show_cm=True):
    scores = ensemble_score(Xs, Xm)
    preds  = (scores > best_t).astype(int)
    levels = predict_with_levels(scores, best_high_t, best_med_t)

    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")

    if len(np.unique(y)) > 1:
        print(classification_report(
            y, preds,
            target_names=["Benign","Ransomware"],
            zero_division=0
        ))
        auc = roc_auc_score(y, scores)
        f1  = f1_score(y, preds, zero_division=0)
        pr  = precision_score(y, preds, zero_division=0)
        rc  = recall_score(y, preds, zero_division=0)
        print(f"  ROC-AUC  : {auc:.4f}")
        print(f"  F1       : {f1:.4f}")
        print(f"  Precision: {pr:.4f}")
        print(f"  Recall   : {rc:.4f}")

        # Alert level distribution
        print(f"\n  Alert levels:")
        print(f"    HIGH   (2): {(levels==2).sum()}")
        print(f"    MEDIUM (1): {(levels==1).sum()}")
        print(f"    Benign (0): {(levels==0).sum()}")

        if show_cm:
            cm = confusion_matrix(y, preds)
            fig, ax = plt.subplots(figsize=(5,4))
            ConfusionMatrixDisplay(
                cm, display_labels=["Benign","Ransomware"]
            ).plot(ax=ax, colorbar=False)
            ax.set_title(f"Confusion Matrix — {name}")
            plt.tight_layout()
            plt.savefig(
                f"/kaggle/working/cm_{name.lower().replace(' ','_')}.png",
                dpi=100
            )
            plt.show()
    else:
        detected = preds.sum()
        total    = len(preds)
        print(f"  Detection rate : {detected}/{total} = {detected/total:.2%}")
        print(f"  Mean score     : {scores.mean():.4f}")
        print(f"  HIGH alerts    : {(levels==2).sum()}")
        print(f"  MEDIUM alerts  : {(levels==1).sum()}")

    return preds, scores


val_preds,  val_s  = evaluate("Validation",          Xs_val, Xm_val, y_val)
ext_preds,  ext_s  = evaluate("Extra (OOD HW)",      Xs_ex,  Xm_ex,  y_extra)
mix_preds,  mix_s  = evaluate("Mix (Simultaneous)",  Xs_mix, Xm_mix, y_mix)
var_preds,  var_s  = evaluate("Variants",            Xs_var, Xm_var, y_variants)


# ── ROC curves ────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7,5))
RocCurveDisplay.from_predictions(y_val,   val_s, name="Validation", ax=ax)
RocCurveDisplay.from_predictions(y_extra, ext_s, name="Extra",      ax=ax)
ax.plot([0,1],[0,1],"k--", alpha=0.5)
ax.set_title("ROC Curves — Council Ensemble")
plt.tight_layout()
plt.savefig("/kaggle/working/roc_council.png", dpi=100)
plt.show()


# ── Score distributions ───────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, (name, scores, y) in zip(axes, [
    ("Validation", val_s, y_val),
    ("Extra",      ext_s, y_extra),
]):
    ax.hist(scores[y==0], bins=50, alpha=0.6,
            color="steelblue", label="Benign",     density=True)
    ax.hist(scores[y==1], bins=50, alpha=0.6,
            color="crimson",   label="Ransomware", density=True)
    ax.axvline(best_med_t,  color="orange", linestyle="--",
               label=f"MEDIUM={best_med_t:.3f}")
    ax.axvline(best_high_t, color="red",    linestyle="--",
               label=f"HIGH={best_high_t:.3f}")
    ax.set_title(f"Score Distribution — {name}")
    ax.legend(fontsize=8)
plt.tight_layout()
plt.savefig("/kaggle/working/score_dist_council.png", dpi=100)
plt.show()


# ── Per-family detection ──────────────────────────────────────
print("\n" + "="*55)
print("PER-FAMILY DETECTION RATE (original split)")
print("="*55)

df_orig  = df[df.split == "original"].copy()
Xs_o     = scaler_s.transform(df_orig[STORAGE_FEATURES].values)
Xm_o     = scaler_m.transform(df_orig[MEMORY_FEATURES].values)
df_orig["score"] = ensemble_score(Xs_o, Xm_o)
df_orig["pred"]  = (df_orig["score"] > best_t).astype(int)

fam = df_orig.groupby(["family","label"]).apply(
    lambda g: pd.Series({
        "total"   : len(g),
        "detected": int(g["pred"].sum()),
        "rate"    : f"{g['pred'].mean():.2%}",
        "mean_score": f"{g['score'].mean():.4f}",
    })
).reset_index()
print(fam.to_string())


# ── Per-model contribution on validation ─────────────────────
print("\n" + "="*55)
print("INDIVIDUAL MODEL AUC ON VALIDATION")
print("="*55)
val_norm_scores = get_normalized_scores(Xs_val, Xm_val)
for name, scores in val_norm_scores.items():
    try:
        auc = roc_auc_score(y_val, scores)
        print(f"  {name:12s}: AUC={auc:.4f}")
    except Exception as e:
        print(f"  {name:12s}: {e}")


# ════════════════════════════════════════════════════════════════
# SAVE EVERYTHING
# ════════════════════════════════════════════════════════════════
joblib.dump({
    "IF_storage":       IF_storage,
    "IF_memory":        IF_memory,
    "ecod":             ecod,
    "copod":            copod,
    "hbos":             hbos,
    "lof":              lof,
    "scaler_storage":   scaler_s,
    "scaler_memory":    scaler_m,
    "normalizers":      normalizers,
    "best_weights":     best_weights,
    "model_names":      model_names,
    "threshold":        best_t,
    "threshold_high":   best_high_t,
    "threshold_medium": best_med_t,
    "storage_features": STORAGE_FEATURES,
    "memory_features":  MEMORY_FEATURES,
}, "/kaggle/working/council_detector.pkl")

print("\n✅ Council saved → /kaggle/working/council_detector.pkl")
```

```python
import numpy as np
import pandas as pd
from pathlib import Path
import pyarrow as pa
import pyarrow.parquet as pq
from tqdm import tqdm
from concurrent.futures import ProcessPoolExecutor

# ── CONFIGURATION ───────────────────────────────────────────
ROOT     = Path("/kaggle/input/datasets/hiranomanabu/ransmap-2024-ransomware-behavioral-features")
OUT_PATH = "/kaggle/working/ransmap_features_v3.parquet"

RANSOMWARE_PURE     = ["AESCrypt","Conti","Darkside","LockBit","REvil","Ryuk","WannaCry"]
ALL_RANSOMWARE      = RANSOMWARE_PURE + ["Conti_01","Conti_02","Conti_03","Conti_04","Conti_05","Conti_06","Conti_07"]
SPLITS              = ["original","extra","mix","variants"]

# Column Definitions
ATA_READ_COLS  = ["unix_sec","unix_ns","lba","size","dummy1","dummy2"]
ATA_WRITE_COLS = ["unix_sec","unix_ns","lba","size","entropy","dummy"]
MEM_READ_COLS  = ["unix_sec","unix_ns","gpa","dummy1","dummy2","type"]
MEM_WRITE_COLS = ["unix_sec","unix_ns","gpa","size","entropy","type"]
MEM_RW_COLS    = ["unix_sec","unix_ns","gpa","size","entropy","type"]
MEM_EXEC_COLS  = ["unix_sec","unix_ns","gpa","dummy1","dummy2","type"]

# ── HELPER FUNCTIONS ────────────────────────────────────────
def entropy_trend(ent):
    n = len(ent)
    if n < 3: return 0.0
    t = n // 3
    return float(ent[2*t:].mean() - ent[:t].mean())

def entropy_variance_blocks(ent, n_blocks=10):
    n = len(ent)
    if n < n_blocks: return float(ent.std()) if len(ent) else 0.0
    bs = n // n_blocks
    return float(np.std([ent[i*bs:(i+1)*bs].mean() for i in range(n_blocks)]))

def write_size_uniformity(sizes):
    if len(sizes) == 0: return 0.0
    counts = pd.Series(sizes).value_counts().values.astype(float)
    p = counts / counts.sum()
    return float(-np.sum(p * np.log2(p[p > 0])))

def io_acceleration(times):
    n = len(times)
    if n < 4: return 0.0
    mid = n // 2
    dur1 = max(times[mid-1] - times[0], 1e-9)
    dur2 = max(times[-1] - times[mid], 1e-9)
    return float((n-mid)/dur2 - mid/dur1)

def infer_metadata(exec_dir):
    parts = exec_dir.parts
    split, split_idx = "unknown", -1
    for i, p in enumerate(parts):
        if p in SPLITS:
            split, split_idx = p, i
            break
    cpu    = parts[split_idx+1] if split_idx >= 0 else "unknown"
    ram    = parts[split_idx+2] if split_idx >= 0 else "unknown"
    family = parts[split_idx+3] if split_idx >= 0 else "unknown"
    label  = 1 if family in ALL_RANSOMWARE or any(s in RANSOMWARE_PURE for s in family.split("_")) else 0
    return label, split, family, cpu, ram

# ── CORE EXTRACTION ─────────────────────────────────────────
def extract_from_dir(exec_dir):
    def load(fname, cols):
        p = exec_dir / fname
        if p.exists():
            return pd.read_csv(p, header=None, names=cols, engine='pyarrow')
        return pd.DataFrame(columns=cols)

    ata_r  = load("ata_read.csv",   ATA_READ_COLS)
    ata_w  = load("ata_write.csv",  ATA_WRITE_COLS)
    mem_r  = load("mem_read.csv",   MEM_READ_COLS)
    mem_w  = load("mem_write.csv",  MEM_WRITE_COLS)
    mem_rw = load("mem_readwrite.csv", MEM_RW_COLS)
    mem_e  = load("mem_exec.csv",   MEM_EXEC_COLS)

    storage   = pd.concat([ata_r, ata_w], ignore_index=True)
    storage_w = ata_w
    memory_w  = pd.concat([mem_w, mem_rw], ignore_index=True)

    feats = {}
    
    # --- Volume & Entropy ---
    ent = storage_w["entropy"].dropna().values if not storage_w.empty else np.array([])
    feats["total_accesses"]  = len(storage)
    feats["mean_entropy"]    = float(ent.mean()) if len(ent) else 0
    feats["max_entropy"]     = float(ent.max()) if len(ent) else 0
    feats["high_entropy_ratio"] = float((ent > 0.85).mean()) if len(ent) else 0
    feats["entropy_trend"]   = entropy_trend(ent)
    
    # --- Temporal ---
    if not storage.empty:
        times = storage["unix_sec"].values.astype(float)
        dur   = max(times[-1] - times[0], 1)
        ar    = len(storage) / dur
        feats["access_rate"] = float(ar)
        feats["io_acceleration"] = io_acceleration(times)
        feats["duration_sec"] = float(dur)
        diffs = np.diff(times)
        feats["burstiness"] = float(diffs.std()/diffs.mean()) if len(diffs) > 0 and diffs.mean() > 0 else 0
    else:
        ar = 0
        feats["access_rate"] = 0
        feats["io_acceleration"] = 0
        feats["duration_sec"] = 1
        feats["burstiness"] = 0

    # --- Spatial ---
    if not storage.empty:
        lba = storage["lba"].values
        feats["unique_blocks"] = int(len(np.unique(lba)))
        feats["block_range"]   = int(lba.max() - lba.min())
    else:
        feats["unique_blocks"] = 0
        feats["block_range"] = 0

    # --- Group 5 & 6 (The Missing Features) ---
    n_reads, n_writes = len(mem_r), len(memory_w)
    rw_ratio = n_reads / max(n_writes, 1)
    feats["rw_ratio"] = float(rw_ratio)
    feats["rw_ratio_deviation"] = float(abs(rw_ratio - 1.0))
    feats["slow_high_entropy"] = 1.0 if (feats["mean_entropy"] > 0.5 and ar < 2000) else 0.0
    feats["inverse_burstiness"] = 1.0 / max(feats["burstiness"], 1)
    
    # NEW V3 FEATURES:
    w_ent = memory_w["entropy"].dropna().values if not memory_w.empty else np.array([])
    we_mean = float(w_ent.mean()) if len(w_ent) else 0
    
    feats["write_entropy_volume"] = float(we_mean * n_writes)
    feats["entropy_rate"] = float(feats["mean_entropy"] * ar / feats["duration_sec"])

    return feats
    
  
    

# ── PARALLEL EXECUTION WRAPPER ──────────────────────────────
def process_single_dir(exec_dir):
    try:
        label, split, family, cpu, ram = infer_metadata(exec_dir)
        feats = extract_from_dir(exec_dir)
        feats.update({
            "label": label, "split": split, "family": family, 
            "cpu": cpu, "ram": ram, "source": str(exec_dir)
        })
        return feats
    except Exception:
        return None

def run_fast_extraction():
    print("Searching for directories...")
    exec_dirs = [p.parent for p in ROOT.rglob("ata_write.csv")]
    print(f"Found {len(exec_dirs)} directories. Starting Parallel Extraction...")

    # Using ProcessPoolExecutor to bypass Python's GIL and use all CPU cores
    with ProcessPoolExecutor() as executor:
        results = list(tqdm(executor.map(process_single_dir, exec_dirs), total=len(exec_dirs)))

    # Filter fails and convert to DataFrame
    final_records = [r for r in results if r is not None]
    df = pd.DataFrame(final_records)
    
    # Save using high compression parquet
    df.to_parquet(OUT_PATH, index=False, engine='pyarrow')
    print(f"✅ Process Complete! Saved to {OUT_PATH}")
    print(f"Final Shape: {df.shape}")

if __name__ == "__main__":
    run_fast_extraction()
```

```python
import numpy as np
import pandas as pd
import joblib
from sklearn.preprocessing import RobustScaler, MinMaxScaler
from sklearn.ensemble import IsolationForest
from pyod.models.hbos import HBOS
from pyod.models.lof import LOF

# 1. Load the v3 dataset containing the new features
df = pd.read_parquet("/kaggle/working/ransmap_features_v3.parquet")
df = df.replace([np.inf, -np.inf], 0).fillna(0)

# 2. Define the exact feature sets used by the Hybrid model
STORAGE_FEATURES = [
    "total_accesses", "mean_entropy", "max_entropy", "high_entropy_ratio",
    "entropy_trend", "access_rate", "io_acceleration", "burstiness",
    "unique_blocks", "block_range"
]

MEMORY_FEATURES = [
    "rw_ratio", "rw_ratio_deviation", "slow_high_entropy", 
    "inverse_burstiness", "write_entropy_volume", "entropy_rate"
]

ALL_FEATURES = STORAGE_FEATURES + MEMORY_FEATURES

# 3. Prepare Training Data (Original split, excluding Zip)
train_df = df[(df.split == "original") & (df.family != "Zip")]
X_train = train_df[ALL_FEATURES].values

s_idx = list(range(len(STORAGE_FEATURES)))
m_idx = list(range(len(STORAGE_FEATURES), len(ALL_FEATURES)))

# 4. Fit Scalers
scaler_s = RobustScaler().fit(X_train[:, s_idx])
scaler_m = RobustScaler().fit(X_train[:, m_idx])

Xs_tr = scaler_s.transform(X_train[:, s_idx])
Xm_tr = scaler_m.transform(X_train[:, m_idx])
Xa_tr = np.hstack([Xs_tr, Xm_tr])

# 5. Train the Council (Unsupervised Specialists)
print("Training the Council of Models...")
if_storage = IsolationForest(n_estimators=200, contamination=0.05, random_state=42).fit(Xs_tr)
if_memory  = IsolationForest(n_estimators=200, contamination=0.05, random_state=42).fit(Xm_tr)
if_full    = IsolationForest(n_estimators=200, contamination=0.05, random_state=42).fit(Xa_tr)
if_diverse = IsolationForest(n_estimators=200, contamination=0.10, bootstrap=True, random_state=7).fit(Xa_tr)
hbos_mdl   = HBOS(contamination=0.05).fit(Xa_tr)
lof_mdl    = LOF(contamination=0.05, n_jobs=-1).fit(Xa_tr)

# 6. Fit Normalizers for Fusion
def get_raw_scores(Xs, Xm):
    Xa = np.hstack([Xs, Xm])
    return {
        "IF_storage": -if_storage.decision_function(Xs),
        "IF_memory":  -if_memory.decision_function(Xm),
        "IF_full":    -if_full.decision_function(Xa),
        "hbos":        hbos_mdl.decision_scores_,
        "lof":         lof_mdl.decision_scores_,
        "IF_diverse": -if_diverse.decision_function(Xa),
    }

raw_scores = get_raw_scores(Xs_tr, Xm_tr)
normalizers = {}
for name, scores in raw_scores.items():
    normalizers[name] = MinMaxScaler().fit(scores.reshape(-1, 1))

# 7. Save to council_v3.pkl
model_names = ["IF_storage", "IF_memory", "IF_full", "hbos", "lof", "IF_diverse"]
best_weights = [0.15, 0.20, 0.20, 0.15, 0.10, 0.20]

joblib.dump({
    "IF_storage": if_storage, "IF_memory": if_memory, "IF_full": if_full,
    "IF_diverse": if_diverse, "hbos": hbos_mdl, "lof": lof_mdl,
    "scaler_storage": scaler_s, "scaler_memory": scaler_m,
    "normalizers": normalizers, "best_weights": best_weights,
    "model_names": model_names, "storage_features": STORAGE_FEATURES,
    "memory_features": MEMORY_FEATURES, "threshold": 0.25,
    "threshold_high": 0.40, "threshold_medium": 0.30
}, "/kaggle/working/council_v3.pkl")

print("✅ council_v3.pkl created with all features!")
```

```python
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import RobustScaler, MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (
    classification_report, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay,
    RocCurveDisplay, f1_score,
    precision_score, recall_score
)
from xgboost import XGBClassifier

# ── Load council ──────────────────────────────────────────────
council = joblib.load("/kaggle/working/council_v3.pkl")

IF_s        = council["IF_storage"]
IF_m        = council["IF_memory"]
IF_f        = council["IF_full"]
IF_d        = council["IF_diverse"]
hbos        = council["hbos"]
lof         = council["lof"]
scaler_s    = council["scaler_storage"]
scaler_m    = council["scaler_memory"]
normalizers = council["normalizers"]
best_weights= council["best_weights"]
model_names = council["model_names"]
best_t      = council["threshold"]
best_high_t = council["threshold_high"]
best_med_t  = council["threshold_medium"]
STORAGE_FEATURES = council["storage_features"]
MEMORY_FEATURES  = council["memory_features"]
ALL_FEATURES     = STORAGE_FEATURES + MEMORY_FEATURES

# ── Reload data ───────────────────────────────────────────────
df = pd.read_parquet("/kaggle/working/ransmap_features_v3.parquet")
df = df.replace([np.inf, -np.inf], 0)

original = df[df.split == "original"]
extra    = df[df.split == "extra"]
mix      = df[df.split == "mix"]
variants = df[df.split == "variants"]

original_no_zip = original[original.family != "Zip"]

X_orig = original_no_zip[ALL_FEATURES].values
y_orig = original_no_zip["label"].values

X_train, X_val, y_train, y_val = train_test_split(
    X_orig, y_orig, test_size=0.2,
    stratify=y_orig, random_state=42
)

X_extra    = extra[ALL_FEATURES].values;    y_extra    = extra["label"].values
X_mix      = mix[ALL_FEATURES].values;      y_mix      = mix["label"].values
X_variants = variants[ALL_FEATURES].values; y_variants = variants["label"].values

s_idx = list(range(len(STORAGE_FEATURES)))
m_idx = list(range(len(STORAGE_FEATURES), len(ALL_FEATURES)))

def scale(X):
    return (scaler_s.transform(X[:, s_idx]),
            scaler_m.transform(X[:, m_idx]))

Xs_tr,  Xm_tr  = scale(X_train)
Xs_val, Xm_val = scale(X_val)
Xs_ex,  Xm_ex  = scale(X_extra)
Xs_mix, Xm_mix = scale(X_mix)
Xs_var, Xm_var = scale(X_variants)

X_tr_all  = np.hstack([Xs_tr,  Xm_tr])
X_val_all = np.hstack([Xs_val, Xm_val])
X_ex_all  = np.hstack([Xs_ex,  Xm_ex])
X_mx_all  = np.hstack([Xs_mix, Xm_mix])
X_vr_all  = np.hstack([Xs_var, Xm_var])


# ════════════════════════════════════════════════════════════════
# COUNCIL SCORING FUNCTIONS
# ════════════════════════════════════════════════════════════════

def get_raw_scores(Xs, Xm):
    Xa = np.hstack([Xs, Xm])
    return {
        "IF_storage": -IF_s.decision_function(Xs),
        "IF_memory":  -IF_m.decision_function(Xm),
        "IF_full":    -IF_f.decision_function(Xa),
        "hbos":        hbos.decision_function(Xa),
        "lof":         lof.decision_function(Xa),
        "IF_diverse":  -IF_d.decision_function(Xa),
    }

def get_norm_scores(Xs, Xm):
    raw = get_raw_scores(Xs, Xm)
    return {
        name: normalizers[name].transform(
            s.reshape(-1, 1)
        ).flatten()
        for name, s in raw.items()
    }

def council_score(Xs, Xm):
    norm = get_norm_scores(Xs, Xm)
    return sum(best_weights[i] * norm[m]
               for i, m in enumerate(model_names))


# ════════════════════════════════════════════════════════════════
# PHASE 1 — GENERATE PSEUDO LABELS FROM COUNCIL
# ════════════════════════════════════════════════════════════════

# Score the FULL original dataset (including Zip)
df_orig      = df[df.split == "original"].copy()
Xs_all_orig  = scaler_s.transform(df_orig[STORAGE_FEATURES].values)
Xm_all_orig  = scaler_m.transform(df_orig[MEMORY_FEATURES].values)
df_orig["council_score"] = council_score(Xs_all_orig, Xm_all_orig)

# Also score extra split
df_extra     = df[df.split == "extra"].copy()
Xs_all_extra = scaler_s.transform(df_extra[STORAGE_FEATURES].values)
Xm_all_extra = scaler_m.transform(df_extra[MEMORY_FEATURES].values)
df_extra["council_score"] = council_score(Xs_all_extra, Xm_all_extra)

# Pseudo-label thresholds
PSEUDO_RANSOM_T = 0.35   # high confidence ransomware
PSEUDO_BENIGN_T = 0.15   # high confidence benign

def make_pseudo_labels(df_split):
    pseudo = df_split.copy()
    pseudo["pseudo_label"] = -1  # unknown
    pseudo.loc[pseudo["council_score"] >= PSEUDO_RANSOM_T,
               "pseudo_label"] = 1
    pseudo.loc[pseudo["council_score"] <= PSEUDO_BENIGN_T,
               "pseudo_label"] = 0
    return pseudo

pseudo_orig  = make_pseudo_labels(df_orig)
pseudo_extra = make_pseudo_labels(df_extra)
pseudo_all   = pd.concat([pseudo_orig, pseudo_extra], ignore_index=True)

# Keep only high-confidence pseudo-labeled rows
pseudo_labeled = pseudo_all[pseudo_all["pseudo_label"] != -1]

print(f"Pseudo-labeled rows : {len(pseudo_labeled)}")
print(f"  Pseudo ransomware : {(pseudo_labeled.pseudo_label==1).sum()}")
print(f"  Pseudo benign     : {(pseudo_labeled.pseudo_label==0).sum()}")
print(f"  Rejected (unsure) : {(pseudo_all.pseudo_label==-1).sum()}")

# Check pseudo-label accuracy on known labels
known = pseudo_labeled[pseudo_labeled["label"] != -1]
match = (known["pseudo_label"] == known["label"]).mean()
print(f"\nPseudo-label accuracy on known: {match:.2%}")

# Family distribution in pseudo-labeled set
print("\nFamily distribution in pseudo-labeled:")
print(pseudo_labeled.groupby(["family","pseudo_label"]).size()
      .reset_index(name="count").to_string())


# ════════════════════════════════════════════════════════════════
# PHASE 2 — TRAIN XGBOOST ON PSEUDO LABELS
# ════════════════════════════════════════════════════════════════

# Scale pseudo-labeled features
Xs_ps = scaler_s.transform(pseudo_labeled[STORAGE_FEATURES].values)
Xm_ps = scaler_m.transform(pseudo_labeled[MEMORY_FEATURES].values)
X_ps  = np.hstack([Xs_ps, Xm_ps])
y_ps  = pseudo_labeled["pseudo_label"].values.astype(int)

# Also add council score as a feature — it's informative
council_feat_ps = pseudo_labeled["council_score"].values.reshape(-1,1)
X_ps_aug = np.hstack([X_ps, council_feat_ps])

# XGBoost — handles class imbalance with scale_pos_weight
n_ben = (y_ps == 0).sum()
n_ran = (y_ps == 1).sum()
spw   = n_ben / max(n_ran, 1)

xgb = XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=spw,
    use_label_encoder=False,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1,
    verbosity=0,
)

# Cross-validate on pseudo-labeled data
print("\nCross-validating XGBoost on pseudo-labels...")
skf    = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_f1s = []

for fold, (tr_idx, va_idx) in enumerate(skf.split(X_ps_aug, y_ps)):
    xgb_cv = XGBClassifier(
        n_estimators=500, max_depth=6, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        scale_pos_weight=spw, use_label_encoder=False,
        eval_metric="logloss", random_state=42,
        n_jobs=-1, verbosity=0
    )
    xgb_cv.fit(X_ps_aug[tr_idx], y_ps[tr_idx])
    preds = xgb_cv.predict(X_ps_aug[va_idx])
    f1 = f1_score(y_ps[va_idx], preds, zero_division=0)
    cv_f1s.append(f1)
    print(f"  Fold {fold+1}: F1={f1:.4f}")

print(f"  Mean CV F1: {np.mean(cv_f1s):.4f} ± {np.std(cv_f1s):.4f}")

# Train final XGBoost on all pseudo-labeled data
xgb.fit(X_ps_aug, y_ps)
print("✅ XGBoost trained on pseudo-labels")


# ════════════════════════════════════════════════════════════════
# PHASE 3 — HYBRID SCORING (COUNCIL + XGBOOST)
# ════════════════════════════════════════════════════════════════

def hybrid_score(Xs, Xm, council_scores=None):
    """
    Combine council anomaly score with XGBoost probability.
    council_scores: pre-computed to avoid double computation
    """
    if council_scores is None:
        council_scores = council_score(Xs, Xm)

    Xa     = np.hstack([Xs, Xm])
    cs_aug = np.hstack([Xa, council_scores.reshape(-1,1)])
    xgb_prob = xgb.predict_proba(cs_aug)[:, 1]

    # Weighted combination
    return 0.4 * council_scores + 0.6 * xgb_prob


# ── Find best hybrid threshold on validation ──────────────────
cs_val   = council_score(Xs_val, Xm_val)
hy_val   = hybrid_score(Xs_val, Xm_val, cs_val)

best_ht, best_hf1 = 0, 0
for t in np.percentile(hy_val, np.arange(10, 95, 0.5)):
    preds = (hy_val > t).astype(int)
    f1    = f1_score(y_val, preds, zero_division=0)
    if f1 > best_hf1:
        best_hf1, best_ht = f1, t

print(f"\nHybrid threshold : {best_ht:.6f}")
print(f"Hybrid Val F1    : {best_hf1:.4f}")

# High/medium thresholds for hybrid
best_hh_t, best_hm_t = best_ht, best_ht
best_hhp, best_hmr   = 0, 0

for t in np.percentile(hy_val, np.arange(10, 99, 0.5)):
    preds = (hy_val > t).astype(int)
    p = precision_score(y_val, preds, zero_division=0)
    r = recall_score(y_val, preds, zero_division=0)
    if p > best_hhp and r > 0.5:
        best_hhp, best_hh_t = p, t
    if r > best_hmr and p > 0.5:
        best_hmr, best_hm_t  = r, t

print(f"HIGH  threshold  : {best_hh_t:.4f}  precision={best_hhp:.3f}")
print(f"MEDIUM threshold : {best_hm_t:.4f}  recall={best_hmr:.3f}")

def hybrid_predict_levels(scores):
    result = np.zeros(len(scores), dtype=int)
    result[scores > best_hm_t]  = 1
    result[scores > best_hh_t]  = 2
    return result


# ════════════════════════════════════════════════════════════════
# PHASE 4 — FULL EVALUATION
# ════════════════════════════════════════════════════════════════

def evaluate_hybrid(name, Xs, Xm, y, show_cm=True):
    cs     = council_score(Xs, Xm)
    scores = hybrid_score(Xs, Xm, cs)
    preds  = (scores > best_ht).astype(int)
    levels = hybrid_predict_levels(scores)

    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")

    if len(np.unique(y)) > 1:
        print(classification_report(
            y, preds,
            target_names=["Benign","Ransomware"],
            zero_division=0
        ))
        auc = roc_auc_score(y, scores)
        print(f"  AUC      : {auc:.4f}")
        print(f"  F1       : {f1_score(y,preds,zero_division=0):.4f}")
        print(f"  Precision: {precision_score(y,preds,zero_division=0):.4f}")
        print(f"  Recall   : {recall_score(y,preds,zero_division=0):.4f}")
        print(f"  HIGH(2)  : {(levels==2).sum()}")
        print(f"  MEDIUM(1): {(levels==1).sum()}")
        print(f"  Benign(0): {(levels==0).sum()}")

        if show_cm:
            cm = confusion_matrix(y, preds)
            fig, ax = plt.subplots(figsize=(5,4))
            ConfusionMatrixDisplay(
                cm, display_labels=["Benign","Ransomware"]
            ).plot(ax=ax, colorbar=False)
            ax.set_title(f"CM Hybrid — {name}")
            plt.tight_layout()
            plt.savefig(
                f"/kaggle/working/hybrid_cm_{name[:8].replace(' ','_')}.png",
                dpi=100
            )
            plt.show()
    else:
        detected = preds.sum()
        print(f"  Detection : {detected}/{len(preds)} = {detected/len(preds):.2%}")
        print(f"  HIGH(2)   : {(levels==2).sum()}")
        print(f"  MEDIUM(1) : {(levels==1).sum()}")

    return preds, scores


vp, vs   = evaluate_hybrid("Validation",  Xs_val, Xm_val, y_val)
ep, es   = evaluate_hybrid("Extra OOD",   Xs_ex,  Xm_ex,  y_extra)
mp, ms   = evaluate_hybrid("Mix",         Xs_mix, Xm_mix, y_mix)
vrp, vrs = evaluate_hybrid("Variants",    Xs_var, Xm_var, y_variants)


# ── ROC curves ────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(7,5))
RocCurveDisplay.from_predictions(y_val,   vs, name="Validation", ax=ax)
RocCurveDisplay.from_predictions(y_extra, es, name="Extra OOD",  ax=ax)
ax.plot([0,1],[0,1],"k--",alpha=0.4)
ax.set_title("ROC — Hybrid (Council + XGBoost)")
plt.tight_layout()
plt.savefig("/kaggle/working/hybrid_roc.png", dpi=100)
plt.show()


# ── Score distributions ───────────────────────────────────────
fig, axes = plt.subplots(1,2,figsize=(14,5))
for ax,(name,scores,y) in zip(axes,[
    ("Validation", vs, y_val),
    ("Extra OOD",  es, y_extra),
]):
    ax.hist(scores[y==0], bins=50, alpha=0.6,
            color="steelblue", label="Benign", density=True)
    ax.hist(scores[y==1], bins=50, alpha=0.6,
            color="crimson", label="Ransomware", density=True)
    ax.axvline(best_hm_t, color="orange", ls="--",
               label=f"MED={best_hm_t:.3f}")
    ax.axvline(best_hh_t, color="red", ls="--",
               label=f"HIGH={best_hh_t:.3f}")
    ax.set_title(f"Hybrid Scores — {name}")
    ax.legend(fontsize=8)
plt.tight_layout()
plt.savefig("/kaggle/working/hybrid_score_dist.png", dpi=100)
plt.show()


# ── Per-family detection ──────────────────────────────────────
print("\n" + "="*55)
print("PER-FAMILY DETECTION — HYBRID (original split)")
print("="*55)
df_orig2      = df[df.split=="original"].copy()
Xs_o2         = scaler_s.transform(df_orig2[STORAGE_FEATURES].values)
Xm_o2         = scaler_m.transform(df_orig2[MEMORY_FEATURES].values)
cs_o2         = council_score(Xs_o2, Xm_o2)
df_orig2["hybrid_score"] = hybrid_score(Xs_o2, Xm_o2, cs_o2)
df_orig2["pred"]         = (df_orig2["hybrid_score"] > best_ht).astype(int)

fam = df_orig2.groupby(["family","label"]).apply(
    lambda g: pd.Series({
        "total"     : len(g),
        "detected"  : int(g["pred"].sum()),
        "rate"      : f"{g['pred'].mean():.2%}",
        "mean_score": f"{g['hybrid_score'].mean():.4f}",
    })
).reset_index()
print(fam.to_string())


# ── XGBoost feature importance ────────────────────────────────
print("\n" + "="*55)
print("XGBOOST TOP 15 FEATURE IMPORTANCES")
print("="*55)
feat_names = ALL_FEATURES + ["council_score"]
importances = pd.Series(
    xgb.feature_importances_,
    index=feat_names
).sort_values(ascending=False)
print(importances.head(15).to_string())


# ── Save hybrid ───────────────────────────────────────────────
joblib.dump({
    "council":          council,
    "xgb":              xgb,
    "threshold":        best_ht,
    "threshold_high":   best_hh_t,
    "threshold_medium": best_hm_t,
    "storage_features": STORAGE_FEATURES,
    "memory_features":  MEMORY_FEATURES,
}, "/kaggle/working/hybrid_detector.pkl")

print("\n✅ Hybrid saved → /kaggle/working/hybrid_detector.pkl")
```

```python
# Remove council_score from XGBoost features
# Force it to learn independently from raw features only

X_ps_raw = X_ps  # no council_score appended

xgb_independent = XGBClassifier(
    n_estimators=500,
    max_depth=4,          # shallower to reduce overfitting
    learning_rate=0.05,
    subsample=0.7,
    colsample_bytree=0.7,
    min_child_weight=5,   # regularization
    scale_pos_weight=spw,
    use_label_encoder=False,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1,
    verbosity=0
)

xgb_independent.fit(X_ps_raw, y_ps)

# New hybrid without circular dependency
def hybrid_score_v2(Xs, Xm):
    cs   = council_score(Xs, Xm)
    Xa   = np.hstack([Xs, Xm])
    prob = xgb_independent.predict_proba(Xa)[:, 1]
    return 0.4 * cs + 0.6 * prob

# Recalibrate threshold
hy_val_v2 = hybrid_score_v2(Xs_val, Xm_val)
best_ht2, best_hf1_2 = 0, 0
for t in np.percentile(hy_val_v2, np.arange(10, 95, 0.5)):
    preds = (hy_val_v2 > t).astype(int)
    f1    = f1_score(y_val, preds, zero_division=0)
    if f1 > best_hf1_2:
        best_hf1_2, best_ht2 = f1, t

print(f"Independent XGBoost Val F1: {best_hf1_2:.4f}")

# Evaluate per family
df_orig3 = df[df.split=="original"].copy()
Xs_o3    = scaler_s.transform(df_orig3[STORAGE_FEATURES].values)
Xm_o3    = scaler_m.transform(df_orig3[MEMORY_FEATURES].values)
df_orig3["hybrid_v2_score"] = hybrid_score_v2(Xs_o3, Xm_o3)
df_orig3["pred"] = (df_orig3["hybrid_v2_score"] > best_ht2).astype(int)

fam2 = df_orig3.groupby(["family","label"]).apply(
    lambda g: pd.Series({
        "total"     : len(g),
        "detected"  : int(g["pred"].sum()),
        "rate"      : f"{g['pred'].mean():.2%}",
        "mean_score": f"{g['hybrid_v2_score'].mean():.4f}",
    })
).reset_index()
print(fam2.to_string())

# Feature importance without council_score
feat_imp2 = pd.Series(
    xgb_independent.feature_importances_,
    index=ALL_FEATURES
).sort_values(ascending=False)
print("\nTop 15 features (no council_score):")
print(feat_imp2.head(15).to_string())
```
